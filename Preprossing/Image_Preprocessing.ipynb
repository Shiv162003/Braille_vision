# Loading the image Using CV2


```python
image = cv2.imread('test_img.jpg')
```

### Normally Cv2 takes image in the form of a BGR image

### BGR (Blue, Green, Red):

BGR is the default color space for OpenCV and many other computer vision libraries. In BGR images, the color channels are arranged in the order Blue, Green, Red. This ordering is historical and comes from the way color data was stored in memory in early computer graphics systems and the format used by certain camera sensors.

### RGB (Red, Green, Blue):

RGB is the standard color space used in most imaging systems and devices, including monitors, cameras, and image processing software. In RGB images, the color channels are arranged in the order Red, Green, Blue. This ordering is more intuitive, as it follows the natural order of primary colors.

```python
# Convert BGR to RGB (matplotlib uses RGB)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
# Matplot lib normally used to display different grphs we dont need axis
plt.axis('off')  
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the grayscale image using matplotlib
plt.imshow(gray_image, cmap='gray')
plt.axis('off')  # Turn off axis numbers and ticks
plt.show()
```

```python
# Apply Gaussian blur
gaussian_blur = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Apply median filtering
median_filtered = cv2.medianBlur(gray_image, 5)

# Calculate histograms
hist_original = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
hist_gaussian = cv2.calcHist([gaussian_blur], [0], None, [256], [0, 256])
hist_median = cv2.calcHist([median_filtered], [0], None, [256], [0, 256])

# Plot histograms
plt.figure(figsize=(10, 6))
plt.plot(hist_original, color='black', label='Original')
plt.plot(hist_gaussian, color='blue', label='Gaussian Blur')
plt.plot(hist_median, color='red', label='Median Filter')
plt.title('Histogram Comparison')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur
gaussian_blur = cv2.GaussianBlur(gray_image, (5, 5), 0)

# Apply contrast enhancement (histogram equalization) to the Gaussian blurred image
equalized_image = cv2.equalizeHist(gaussian_blur)

# Calculate histograms
hist_original = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
hist_gaussian = cv2.calcHist([gaussian_blur], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(gaussian_blur, cmap='gray')
plt.title('Gaussian Blur')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalization')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.plot(hist_original, color='black', label='Original')
plt.plot(hist_gaussian, color='blue', label='Gaussian Blur')
plt.plot(hist_equalized, color='red', label='Histogram Equalization')
plt.title('Histogram Comparison')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```
